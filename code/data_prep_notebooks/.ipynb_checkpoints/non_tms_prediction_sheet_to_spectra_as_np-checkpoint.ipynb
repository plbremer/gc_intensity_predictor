```python
def parse_in_silico_string_to_np_spectrum(temp_string):
    this_spectrum_list=list()
    for mz,element in enumerate(temp_string.split()[1:]):
        #splitting on . and taking the first gives us the number
        #print(element)
        
        try: 
            intensity=float(element.split('.')[0])
        except ValueError:
            continue
    #     print(intensity)
    #     print(mz)

        if mz==0:
            continue
        if intensity>0:
            this_spectrum_list.append(
                [float(mz),intensity]
            )
    #     #print(element.split('.'))
    spectrum_array=np.array(this_spectrum_list)
    spectrum_array[:,1]=spectrum_array[:,1]/spectrum_array[:,1].max()
    return spectrum_array
```

```python
spectrum_np_list=list()
for index,series in non_tms_panda.iterrows():
    spectrum_np_list.append(
        parse_in_silico_string_to_np_spectrum(series['In-silico spectrum'])
    )
    
non_tms_panda['spectrum_np']=spectrum_np_list
```
