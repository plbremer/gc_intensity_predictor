```python
def coerce_one_spectrum_to_mz_position(temp_spectrum):

    #print(temp_spectrum)
    
    output_array=np.zeros(
        501
    )
    
    

    #set the particular mz to 1
    #we set one value in each row to one
    #the mz are the things that we set to one
    output_array[temp_spectrum[:,0].astype(int)]=temp_spectrum[:,1]
    

    return output_array
```

```python
quantum_spectra_panda=quantum_spectra_panda.loc[:,['Short','spectrum_np']]
```

```python
nist_measured_panda=nist_measured_panda.loc[:,['inchikey_first_block','spectrum_np']]
```

```python
quantum_spectra_panda['spectrum_for_similarity_quantum']=quantum_spectra_panda['spectrum_np'].apply(
    coerce_one_spectrum_to_mz_position
)
```

```python
nist_measured_panda['spectrum_for_similarity_experimental']=nist_measured_panda['spectrum_np'].apply(
    coerce_one_spectrum_to_mz_position
)
```

```python
total_panda_for_similarity=nist_measured_panda.merge(
    right=quantum_spectra_panda,
    left_on='inchikey_first_block',
    right_on='Short',
    how='left'
)
```

```python
def crappy_cosine(a,b):
    print(a)
    print(b)
    return dot(a, b)/(norm(a)*norm(b))
```

```python
crappy_cosine(
    total_panda_for_similarity.at[0,'spectrum_for_similarity_experimental'],
    total_panda_for_similarity.at[0,'spectrum_for_similarity_quantum']
)
```

```python
from scipy.spatial.distance import cosine
```

```python
cosine(
    total_panda_for_similarity.at[0,'spectrum_for_similarity_experimental'],
    total_panda_for_similarity.at[0,'spectrum_for_similarity_quantum']
)
```

```python
cosine_sim_list=list()
for index,series in total_panda_for_similarity.iterrows():
    cosine_sim_list.append(
        cosine(
            series['spectrum_for_similarity_experimental'],
            series['spectrum_for_similarity_quantum']
        )
    )
total_panda_for_similarity['cosine']=cosine_sim_list
```
